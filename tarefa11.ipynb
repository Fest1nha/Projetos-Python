```python
# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, transpile
from qiskit.visualization import plot_histogram, plot_bloch_multivector, visualize_transition, circuit_drawer
import numpy as np
from ibm_quantum_widgets import *
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector, random_statevector
backend = AerSimulator()

# qiskit-ibmq-provider has been deprecated.
# Please see the Migration Guides in https://ibm.biz/provider_migration_guide for more detail.
from qiskit_ibm_runtime import QiskitRuntimeService, Sampler, Estimator, Session, Options

# Loading your IBM Quantum account(s)
service = QiskitRuntimeService(channel="ibm_quantum")
```

```python
qc = QuantumCircuit(1)

# Parâmetros aleatórios para as rotações
theta_x = np.random.uniform(0, 2*np.pi)
theta_y = np.random.uniform(0, 2*np.pi)
theta_z = np.random.uniform(0, 2*np.pi)

# Aplicando rotação em torno do eixo x
qc.rx(theta_x, 0)

# Aplicando rotação em torno do eixo y
#qc.ry(theta_y, 0)

# Aplicando rotação em torno do eixo z
qc.rz(theta_z, 0)

#Salvando e medindo para plotar na esfera de bloch
qc.save_statevector()
qc.measure_all()

#Mostrando o Circuito
qc.draw()
```

```python
job = backend.run(qc)
result = job.result()
vetor = result.get_statevector()

plot_bloch_multivector(vetor)
```
